```python
import matplotlib
import pandas as pd
import seaborn as sns
```

```python
#Load CSV
figures_dataframe = pd.read_csv("./results/figures_dataset.txt",sep=" ")
figures_dataframe.head()
```

```python
# Use the 'hue' argument to provide a factor variable
plot = sns.lmplot( x="ph1", y="ph2", data=figures_dataframe, fit_reg=False, hue='Figure_name', legend=True)
 
# Move the legend to an empty part of the plot
plot.savefig("ph1_ph2_plot.png")

```
